```python
from statistics import *
```

```python
#listas de calificaciónes dek grupo a de probabilidad y estadisitica

grupoa= [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print (grupoa)

#sacar la medua de las calificaciones con formula

promedio = sum(grupoa) / len(grupoa)
print (promedio)

#utiliza la funcion mean que esta en la libreria statistic para sacr el promedio

promedio2 = mean(grupoa)
print(promedio2)
```

```python
#determinar el valor de la meduana con la funcion median
mediana = median(grupoa)
print (mediana)
```

```python
# Determinar el valor de la moda con la funcion mode

moda = mode(grupoa)
print (moda)
```

```python
# en caso de que exusta más de una moda, se puede emplear la funcion multimode para multipes modas que puedan existir

moda2 = multimode(grupoa)
print (moda2)
```
