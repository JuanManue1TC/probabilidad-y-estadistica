```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Cargar el Dataset de Autos
autos = pd.read_csv("autos2.csv", encoding="ISO-8859-1")

# Ver información del dataset
autos.info()

# Estadísticas descriptivas básicas
print("\nEstadísticas descriptivas de Price:")
print(autos.price.describe())

# Calcular medidas de tendencia central
top_price = autos.price.max()
low_price = autos.price.min()
mode_price = autos.price.mode()
mean_price = autos.price.mean()
median_price = autos.price.median()
std_price = autos.price.std()

print("\nValores estadísticos:")
print(f"Precio máximo: {top_price}")
print(f"Precio mínimo: {low_price}")
print(f"Moda de precios: {mode_price.values}")
print(f"Media de precios: {mean_price:.2f}")
print(f"Mediana de precios: {median_price:.2f}")
print(f"Desviación estándar: {std_price:.2f}")

# Histograma de precios (Regla de Sturges)
n = len(autos.price.dropna())  # Eliminar valores nulos antes de calcular
k = int(1 + np.log2(n))

plt.figure(figsize=(8, 5))
plt.hist(autos.price.dropna(), bins=k, color='skyblue', edgecolor='black')
plt.title("Histograma de precios")
plt.axvline(mean_price, color="y", linestyle="dashed", linewidth=2, label=f'Media={mean_price:.2f}')
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

# Diagramas de caja y bigotes para analizar relación entre Price y diferentes categorías
plt.figure(figsize=(10, 5))
sns.boxplot(x="price", y="body-style", data=autos, palette="coolwarm")
plt.title("Precio vs. Body Style")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x="price", y="drive-wheels", data=autos, palette="coolwarm")
plt.title("Precio vs. Drive Wheels")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x="price", y="engine-location", data=autos, palette="coolwarm")
plt.title("Precio vs. Engine Location")
plt.show()

```

Conclusión:
Los precios de los autos varían según su estilo, tracción y ubicación del motor. Los convertibles y hardtops son más caros, mientras que los hatchbacks tienen mayor variabilidad. Los vehículos con tracción trasera y motor trasero tienden a ser más costosos, indicando que estos factores influyen directamente en el precio.
