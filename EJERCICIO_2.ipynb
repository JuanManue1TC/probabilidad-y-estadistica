```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros iniciales
media = 70
desv_std = 4
num_datos = 500

# Generar datos
datos = np.random.normal(media, desv_std, num_datos)
print(datos[0:50])

# Histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribucin Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()

# Funcin de densidad de probabilidad (PDF)
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)
pdf = norm.pdf(x, media, desv_std)

print("Primeros 50 valores de x:", x[0:50])
print("Primeros 50 valores de pdf:", pdf[0:50])

# Graficar PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label=f'Normal:(media:{media}; ds:{desv_std})')
plt.title('Funcin de densidad de probabilidad (PDF)')
plt.xlabel('Peso en kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# Histograma con PDF superpuesta
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2)
plt.title('Distribucin Normal con PDF')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()

# Ejercicio 2: Probabilidad de pesar menos de 75 kg
peso_limite = 75
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_75kg = num_datos * prob_menos_75kg

print("Probabilidad < 75 kg:", prob_menos_75kg)
print("Estudiantes < 75 kg:", estudiantes_menos_75kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg\nX = 500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()

# Ejercicio: Probabilidad de pesar menos de 62 kg
peso_limite = 62
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_62kg = num_datos * prob_menos_62kg

print("Probabilidad < 62 kg:", prob_menos_62kg)
print("Estudiantes < 62 kg:", estudiantes_menos_62kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, num_datos)
y = norm.pdf(x, media, desv_std)

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg\nX = 500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()

# Ejercicio: Probabilidad de pesar ms de 79 kg
peso_limite = 79
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)
prob_mas_79kg = 1 - prob_menos_79kg
estudiantes_mas_79kg = num_datos * prob_mas_79kg

print("Probabilidad > 79 kg:", prob_mas_79kg)
print("Estudiantes > 79 kg:", estudiantes_mas_79kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, num_datos)
y = norm.pdf(x, media, desv_std)

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese ms de 79 kg\nX = 500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
